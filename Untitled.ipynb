```python
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;
import seaborn as sns;
```

```python
trajFile = "leapFrog.txt";
N = 3;
names = [];
for i in range(N) :
    names.extend(["rx" + str(i), "ry" + str(i), "rz" + str(i)])
names.append("E");
```

```python
df = pd.read_csv(trajFile, delimiter=" ", names=names)
display(df.head())

df.loc[0]["E"]
np.max(np.abs(df["E"]))
```
