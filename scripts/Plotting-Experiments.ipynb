```python
serialFile = "NBody-Serial-0.01-10-0.5.txt";
file2 = "NBody-2-0.01-10-0.5.txt";
file4 = "NBody-4-0.01-10-0.5.txt";
file8 = "NBody-8-0.01-10-0.5.txt";
file12 = "NBody-12-0.01-10-0.5.txt";
file16 = "NBody-16-0.01-10-0.5_trial2.txt"; #"NBody-16-0.01-10-0.5.txt";
file24 = "NBody-24-0.01-10-0.5.txt";
```
