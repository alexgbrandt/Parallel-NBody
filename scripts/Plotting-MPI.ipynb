```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
serialFile = "Results/mpi-noviz-test_1_-0.01-5-0.5.txt";
file2 = "Results/mpi-noviz-test-2x-0.01-5-0.5.txt";
file4 = "Results/mpi-noviz-test-4x-0.01-5-0.5.txt";
file8 = "Results/mpi-noviz-test-8x_1_-0.01-5-0.5.txt";
file10 = "Results/mpi-noviz-test-10x-_1_0.01-5-0.5.txt";
file20 = "Results/mpi-noviz-test-20x-0.01-5-0.5.txt";
file40 = "Results/mpi-noviz-test-40x-0.01-5-0.5.txt";
file80 = "Results/mpi-noviz-test-80x-0.01-5-0.5.txt";
file100 = "Results/mpi-noviz-test-100x-0.01-5-0.5.txt";
```

```python
dfSerial = pd.read_csv(serialFile, comment="#")
dfSerial['p'] = 1
df2 = pd.read_csv(file2, comment="#");
df2['p'] = 2
df4 = pd.read_csv(file4, comment="#");
df4['p'] = 4
df8 = pd.read_csv(file8, comment="#");
df8['p'] = 8
df10 = pd.read_csv(file10, comment="#");
df10['p'] = 10
df20 = pd.read_csv(file20, comment="#");
df20['p'] = 20
df40 = pd.read_csv(file40, comment="#");
df40['p'] = 40
df80 = pd.read_csv(file80, comment="#");
df80['p'] = 80
df100 = pd.read_csv(file100, comment="#");
df100['p'] = 100
# display(dfSerial.head())
# display(df2.columns)
```

```python
alg0 = pd.DataFrame();
for (d,p) in zip([dfSerial,df2,df4,df8,df10,df20,df40,df80,df100],[1,2,4,8,10,20,40,80,100]):
    alg0 = alg0.append(d[d['Alg']==0], ignore_index=1);

alg1 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg1 = alg1.append(d[d['Alg']==1], ignore_index=1);

    
alg2 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg2 = alg2.append(d[d['Alg']==2], ignore_index=1);

alg3 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg3 = alg3.append(d[d['Alg']==3], ignore_index=1);

alg4 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg4 = alg4.append(d[d['Alg']==4], ignore_index=1);

alg5 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg5 = alg5.append(d[d['Alg']==5], ignore_index=1);

alg6 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg6 = alg6.append(d[d['Alg']==6], ignore_index=1);

alg7 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    alg7 = alg7.append(d[d['Alg']==7], ignore_index=1);
```

```python
N80 = pd.DataFrame();
for d in [dfSerial,df2,df4,df8,df10,df20,df40,df80,df100] :
    N80 = N80.append(d[(d['p']==1) | (d['p']==100)], ignore_index=1);
    
N80['dE'] = np.abs(N80['E0'] - N80['Eend'])/np.abs(N80['E0']) * 100
N80.drop(columns=['E0','T','Eend'],inplace=True)
N80.set_index(['Alg','p'],inplace=True)
N80.fillna(0,inplace=True)
# display(N80.head(20))
pt = N80.pivot_table(values='dE',index=['Alg','p'],columns='N')
# print(pt.to_latex(float_format="%.4f"))
```

# Foreach algorithm:
## Exec time vs N, foreach p
## Speedup vs p, foreach N


```python
ftsize = 8;
aspect = 1.0;

w, h = plt.figaspect(aspect);
fig,axes = plt.subplots(3,3,figsize=(3*w,3*h),dpi=1000);

    
# for algDF in [alg0,alg1,alg2,alg3,alg4,alg5,alg6,alg7] :
for (algDF,ax1,title) in zip([alg1,alg5,alg7],axes[0],["Alg. 1", "Alg. 5", "Alg. 7"]) :

    l = list(set(algDF[algDF['N'] < 100000]['N']));
    l.sort()
    for N in l :
        d = algDF[algDF["N"]==N];

        ax1.plot(d["p"], d[d['p']==1]['T'].values/d['T'], label=str(N));#,linewidth=w,marker=m,markersize=s, fillstyle=f,markeredgecolor='k',markeredgewidth=0.25);

    ax1.axline([0,0],[1,1], color="#555555", linestyle='--', linewidth=0.75, label='ideal')

    ax1.set_title("Parallel Speed-up vs Num. Processes, " + title,fontsize=ftsize+2);
    ax1.set_xlabel("Num. Processes ($p$)",fontsize=ftsize+1);
    
    if (ax1 == axes[0][0]) :
        ax1.set_ylabel("Parallel Speed-up ($\\frac{T_1}{T_p}$) ",fontsize=ftsize+1)
    ax1.tick_params(labelsize=ftsize-1);
    ax1.set_xlim(0,100);
    ax1.set_ylim(0,40);
    ax1.legend(fontsize=ftsize-1, title="$N$",title_fontsize=ftsize,loc='upper left')
    ax1.grid(axis='y', linestyle="-", color='k', linewidth=0.3, alpha=0.4)



# for algDF in [alg0,alg1,alg2,alg3,alg4,alg5,alg6,alg7] :
for (algDF,ax1,title) in zip([alg1,alg5,alg7],axes[1],["Alg. 1", "Alg. 5", "Alg. 7"]) :

    markerlist = ['p', '^','P','s','*', 'X', 'D', 'o','v' ]
    for p,m in zip([1,2,4,8,10,20,40,80,100],markerlist) :
        d = algDF[algDF['p']==p];
        ax1.plot(d["N"], d['T'],label=str(p));#,linewidth=w,marker=m,markersize=s, fillstyle=f,markeredgecolor='k',markeredgewidth=0.25);

    
    ax1.tick_params(labelsize=ftsize-1);
    ax1.set_xscale('log')
    ax1.set_xlim(9000, 85000);
    ax1.set_ylim(10, 12000);
    ax1.set_xticks(np.arange(1,9)*(10**4));
    ax1.set_xticklabels(np.arange(1,9));
    ax1.set_yscale('log')
    ax1.legend(fontsize=ftsize-1, title="Num. Proc. ($p$)" ,ncol=3,title_fontsize=ftsize,loc='upper left')
    ax1.grid(axis='y', linestyle="-", color='k', linewidth=0.3, alpha=0.4)
    ax1.grid(axis='y',which='minor', linestyle="-", color='k', linewidth=0.15, alpha=0.2)
    
    if (ax1 == axes[1][0]) :
        ax1.set_ylabel("Execution Time (s)",fontsize=ftsize+1)
    ax1.set_xlabel("$N$ ($\\times 10^4$)",fontsize=ftsize+1)

    ax1.set_title("Execution Time vs Simulation Size, " + title,fontsize=ftsize+1.5);

    
for (algDF,ax1,title) in zip([alg1,alg5,alg7],axes[2],["Alg. 1", "Alg. 5", "Alg. 7"]) :

    markerlist = ['p', '^','P','s','*', 'X', 'D', 'o','v' ]
    pl = [1,2,4,8,10,20,40,80,100];
    for p,m in zip(pl,markerlist) :
        d = algDF[algDF['p']==p];
        ax1.plot(d["N"], d['p']*d['T']/d['N'],label=str(p));#,linewidth=w,marker=m,markersize=s, fillstyle=f,markeredgecolor='k',markeredgewidth=0.25);

    ax1.tick_params(labelsize=ftsize-1);
    ax1.set_xscale('log')
    ax1.set_xlim(9000, 85000);
    ax1.set_ylim(0.025, 3.0)
    ax1.set_xticks(np.arange(1,9)*(10**4));
    ax1.set_xticklabels(np.arange(1,9));
    ax1.legend(fontsize=ftsize-1, title="Num. Procs. ($p$)",ncol=3,title_fontsize=ftsize,loc='upper right')
    ax1.tick_params(labelsize=ftsize-1);
    ax1.set_yscale('log')
    ax1.grid(axis='y', linestyle="-", color='k', linewidth=0.3, alpha=0.4)
    ax1.grid(axis='y',which='minor', linestyle="-", color='k', linewidth=0.15, alpha=0.2)

    ax1.set_xlabel("$N$ ($\\times10^4$)",fontsize=ftsize+1);
    if (ax1==axes[2][0]) :
        ax1.set_ylabel("$pT_p\; /\; N$ (s)",fontsize=ftsize+1)

    ax1.set_title("Parallel Cost per Body vs $N$, " + title,fontsize=ftsize+1.5);
    

    
plt.tight_layout()
plt.subplots_adjust(hspace=0.25)
    
```
